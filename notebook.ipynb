```python
import pandas as pd
import matplotlib.pyplot as plt

netflix_df = pd.read_csv("./netflix_data.csv", index_col = 0)
```

```python
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]

# Qual a duração mais frequentes dos filmes em 1990?
movies_1990s = netflix_subset[(netflix_subset["release_year"] >= 1990) & (netflix_subset["release_year"] < 2000)]
```

```python
# Vizualizando os daodos
plt.hist(movies_1990s['duration'])
plt.title("Distribuição da duração dos filmes em 1990")
plt.xlabel("Duração (em min)")
plt.ylabel("Número de filmes")
```

```python
# Quantos filmes de ação podem ser considerados curtametragem? (duração menor que 90min)
short_movie_count = movies_1990s[(movies_1990s["duration"] < 90) & (movies_1990s["genre"] == "Action")]
count = len(short_movie_count)
print(count)
```
